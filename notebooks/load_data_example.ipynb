```python
import pandas as pd
```

```python
data = "../data/data.json"

import json

with open(data, 'r') as data_file:
    json_data = data_file.read()

input_data = json.loads(json_data)
```

```python
card_id = "cardId"

main_fields = [
    'cardType',
    'cardSubType',
    'status',
]

location_fields = [
    "district",
    "city",
]

data_fields = [
 'rooms',
 'roomConfiguration',
 'price',
 'size',
 'buildYear',
 'sizeLot',
 'sizeMin',
 'sizeMax',
 'newDevelopment',
 'maintenanceFee',
 'floor',
 'buildingFloorCount',
 'pricePerSqm',
 'condition',
 'sourceType',
]
```

```python
data = {}


for page in input_data:
    for ad in page:
        print(ad)
        print(ad['cardId'])
        uid = ad[card_id]
        
        data[uid] = {}

        for field in main_fields:
            data[uid][field] = ad[field]
        for field in location_fields:
            data[uid][field] = ad["location"][field]
        for field in data_fields:
            data[uid][field] = ad["data"][field]
```

```python
df = pd.DataFrame(data).T
```
